```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(os.path.join('data.csv'))
data.head()
```

```python
data.describe(include='all')
```

```python
target = data['Lithology (major)']
target_names = ['sandstone', 'siltstone', 'shale', 'coal', 'unspecified', 'limestone',
                'dolomite', 'salt', 'volcanic rock', 'metamorphic rock', 'claystone']
```

```python
data = data.drop(columns=['State', 'Well Name', 'Depth (m)', 'Permeability_v (md)', 'Well Classification',
                          'Longitude', 'Latitude', 'Onshore/Offshore', 'Decision'])
data.head()
```

```python
data = data.drop('Lithology (major)', axis=1)
feature_names = data.columns
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f'k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}')
```

```python
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker='x')
plt.xlabel('k neighbors')
plt.ylabel('Testing Accuracy Score')
plt.savefig("images/testing_accuracy_score.png")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print('k=3 Test Accuracy: %.3f' % knn.score(X_test, y_test))
```

```python
new_data = [[0, 0], [10, 100], [50, 500]]
predicted_class = knn.predict(new_data)
print(predicted_class)
```
